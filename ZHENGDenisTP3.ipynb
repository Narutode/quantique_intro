Imports

```python
from qiskit import QuantumCircuit, ClassicalRegister, QuantumRegister
```

Initialisation du circuit quantique avec 10 qubits

```python
q = QuantumRegister(10)
c = ClassicalRegister(1)
circuit = QuantumCircuit(q, c)
```

```python
for i in range(4):
    circuit.h(i)

circuit.barrier()
circuit.draw('mpl')
```

```python
circuit.h(4)
circuit.h(5)

circuit.barrier()
circuit.draw('mpl')
```

```python
circuit.cx(1, 6)
circuit.cx(3, 6)

circuit.barrier()
circuit.draw('mpl')
```

```python
circuit.cx(2, 7)
circuit.cx(3, 7)

circuit.barrier()
circuit.draw('mpl')
```

```python
circuit.cx(4, 6)
circuit.cx(5, 7)

circuit.barrier()
circuit.draw('mpl')
```

```python
circuit.ccx(7, 6, 3)
circuit.barrier()
circuit.x(6)
circuit.ccx(7, 6, 2)
circuit.x(6)
circuit.barrier()
circuit.x(7)
circuit.ccx(7, 6, 1)
circuit.x(7)
circuit.barrier()
circuit.x(6)
circuit.x(7)
circuit.ccx(7, 6, 0)
circuit.x(7)
circuit.x(6)

circuit.barrier()circuit.draw('mpl')
```
